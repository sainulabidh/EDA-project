```python
#removing out layres
import numpy as np
x=['engine_capacity cc','kms_driven','max_power_in_bhp','mileage',]
for i in x:
    data=df1[i]
    q1=np.quantile(data,0.25)
    q2=np.quantile(data,0.50)
    q3=np.quantile(data,0.75)
    iqr=q3-q1
    lower=q1-(1.5*iqr)
    upper=q3+(1.5*iqr)
    outliers=[j for j in data if j<lower or j>upper]
    print(len(outliers))
    df1= df1[(df1[i] >= lower) & (df1[i] <= upper)]
```

```python
#model evaluation
from sklearn.model_selection import train_test_split
y=df1['resale_price_in_lakhs'] # extracting target variable
x=df1.drop(columns=['resale_price_in_lakhs']) # creat training variable by droping
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
#print(f'x_train: {x_train}\n,y_train: {y_train}\n,x_test: {x_test}\n,y_test: {y_test}')
print(f'x_train: {x_train.shape}\ny_train: {y_train.shape}\nx_test: {x_test.shape}\ny_test: {y_test.shape}')
```

```python
print(x_train)
y_train
#print(x_test)
y_test
```

```python
#import numpy as n
from sklearn.svm import SVR
svm_model=SVR()
#fit the model to the train data
svm_model.fit(x_train,y_train)
#predict the target value for test data
svm_prediction=svm_model.predict(x_test)
print('SVM_PREDICTION',svm_prediction) #y_test():-predicted value
```

```python
#calculating error
from sklearn.metrics import mean_absolute_error
svm_mae=mean_absolute_error(y_test,svm_prediction)
print('SVM mean absolute error:',svm_mae)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
svm_mse=mean_squared_error(y_test,svm_prediction)
svm_rmse=sqrt(svm_mse)
print('SVM ROOT MEAN SQUARED ERROR: ',svm_rmse)
```

```python
from sklearn.metrics import r2_score
r=r2_score(y_test,svm_prediction)
r
```
