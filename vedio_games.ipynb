The dataset is a video game sales dataset with 16,598 rows and 11 columns. Here is an overview of its structure:

Columns:
* Rank: Integer, the ranking of the game based on global sales.
* Name: Object, the name of the video game.
* Platform: Object, the gaming platform (e.g., Wii, NES, PS).
* Year: Float64, the year of release.
* Genre: Object, the game's genre (e.g., Sports, Platform, Racing).
* Publisher: Object, the company that published the game.
* NA_Sales: Float64, sales in North America (in millions of units).
* EU_Sales: Float64, sales in Europe (in millions of units).
* JP_Sales: Float64, sales in Japan (in millions of units).
* Other_Sales: Float64, sales in other regions (in millions of units).
* Global_Sales: Float64, total global sales (in millions of units).


* Dataset Details:
* Shape: 16,598 rows Ã— 11 columns
* Data Types: A mix of numerical (integer, float) and categorical (object) columns.

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns='Global_Sales')
y=df['Global_Sales']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
```

```python
#importing model to evaluate
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#predicting the data using x_test
model.fit(x_train,y_train)
prediction=model.predict(x_test)
```

```python
# finding mae
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,prediction)
mae
```

```python
#calculating mse
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,prediction)
mse
```

```python
#calculating rmse
from math import sqrt
rmse=sqrt(mse)
rmse
```

```python
#calculating the accuracy of the model
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,prediction)
accuracy
```
